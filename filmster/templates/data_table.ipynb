```python
import os
import csv
import pandas as pd
```

```python
data_file = "static\\data\\df_sub.csv"
```

```python
df = pd.read_csv(data_file)
df.head()
```

```python
df2 = df.groupby(['title']).mean()
df2
```

```python
df3 = df.groupby(['title']).count()
df3
```

```python
df_genres = df[['title', 'genres']].drop_duplicates(subset='title', keep='first')
df_genres
```

```python
merge_df = pd.merge(df2, df_genres, on="title")
```

```python
merge_df2= pd.merge(df3, merge_df, on="title")
merge_df2
```

```python
df_clean = merge_df2[['title','user_id_x', 'rating_y', 'age_y', 'genres_y' ]]
```

```python
df_clean
```

```python
df_table = df_clean.rename(columns={'user_id_x': 'user_count', 'rating_y': 'rating_avg','age_y':'user_age_avg', 'genres_y':'genres'})
```

```python
df_table
```

```python
df_table['genres'] = df_table['genres'].str.replace('|',', ')
```

```python
df_table
```

```python
df_table = df_table[['title', 'genres','user_count','rating_avg','user_age_avg']]
df_table
```

```python
df_table['rating_avg'] = df_table['rating_avg'].round(decimals=2)
```

```python
df_table['user_age_avg'] = df_table['user_age_avg'].round(decimals=2)
```

```python
df_table
```

```python
df_table.to_csv("static\\data\\Table_Data.csv", index=False, header=True)
```
